# 5. Cricket rankings for men
### (a) Top 10 ODI Teams

```python
page = requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
page
```

```python
soup = bs(page.content, 'html.parser')
#soup
```

#### matches, points and rating for no.1 team

```python
team_1 = []
for i in soup.find_all('tr', class_="rankings-block__banner"):
    team_1.append(i.find(class_="u-hide-phablet").text)

match_1 = []
for i in soup.find_all('td',class_="rankings-block__banner--matches"):
    match_1.append(i.text)

points_1 = []
for i in soup.find_all('td',class_="rankings-block__banner--points"):
    points_1.append(i.text)

rating_1 = []
for i in soup.find_all('td',class_="rankings-block__banner--rating u-text-right"):
    rating_1.append(i.text.replace("\n","").strip("   "))
```

#### matches, points and rating for other teams

```python
teams = []
for i in soup.find_all('tr',class_="table-body"):
    teams.append(i.find('span',class_ = 'u-hide-phablet').text)
teams = teams[:9]

matches = []
for i in soup.find_all('tr',class_="table-body"):
    matches.append(i.find('td', class_ = "table-body__cell u-center-text").text)
    
points = []
for i in soup.find_all('td', class_ = "table-body__cell u-center-text"):
    points.append(i.text)
points = points[1::2]

ratings = []
for i in soup.find_all('td', class_="table-body__cell u-text-right rating"):
    ratings.append(i.text)
```

#### Matches, Points, Ratings for all Teams

```python
Teams = team_1 + teams

Matches = match_1 + matches
Matches = Matches[:10]

Points = points_1 + points
Points = Points [:10]

Ratings = rating_1 + ratings
Ratings = Ratings[:10]
```

```python
df = pd.DataFrame()
df['Teams'] = Teams
df['Matches'] = Matches
df['Points'] = Points
df['Ratings'] = Ratings
df
```

# 9. Restaurant details

```python
page = requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
page
```

```python
soup = bs(page.content,'html.parser')
#soup
```

```python
name = []
for i in soup.find_all('div',class_="restnt-info cursor"):
    name.append(i.find('a').text)

cuisine = []
for i in soup.find_all('span',class_="double-line-ellipsis"):
    cuisine.append(i.find('a').text)

location = []
for i in soup.find_all('div',class_="restnt-loc ellipsis"):
    location.append(i.text)

ratings = []
for i in soup.find_all('div',class_="restnt-rating rating-4"):
    ratings.append(i.text)

image = []
for i in soup.find_all('img', class_ = 'no-img'):
    image.append(i['data-src'])
```

```python
df = pd.DataFrame()
df["Restaurant name"] = name
df['Cuisine'] = cuisine
df['Location'] = location
df['Ratings'] = ratings
df['Image url'] = image
df
```

# 10. Top publications from Google Scholar

```python
page = requests.get('https://scholar.google.com/citations?view_op=top_venues&hl=en')
page
```

```python
soup = bs(page.content,'html.parser')
#soup
```

```python
rank = []
for i in soup.find_all('td',class_="gsc_mvt_p"):
    rank.append(i.text)

publication = []
for i in soup.find_all('td',class_="gsc_mvt_t"):
    publication.append(i.text)

h5_index = []
for i in soup.find_all('a',class_="gs_ibl gsc_mp_anchor"):
    h5_index.append(i.text)

h5_median = []
for i in soup.find_all('span',class_="gs_ibl gsc_mp_anchor"):
    h5_median.append(i.text)
```

```python
df = pd.DataFrame()
df["Rank"] = rank
df["Publication"]  = publication
df["h5-index"] = h5_index
df["h5-median"] = h5_median
df
```
